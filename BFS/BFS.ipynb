```python
import networkx as nx
import matplotlib.pyplot as plt
import collections
```

```python
# Define the graph as an adjacency set
graph = {'0': {'1', '2'},
         '1': {'0', '3', '4'},
         '2': {'0'},
         '3': {'1'},
         '4': {'2', '3'}}
```

```python
# Function to draw adjacency matrix
def draw_adjacency_matrix(graph):
    G = nx.Graph(graph)
    A = nx.adjacency_matrix(G).todense()
    plt.matshow(A, cmap='viridis')
    plt.title("Adjacency Matrix")
    plt.colorbar()
    plt.show()
```

```python
# Function to draw the graph
def draw_graph(graph):
    G = nx.Graph(graph)
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=1500, font_size=10, font_color='black')
    plt.title("Graph")
    plt.show()
```

```python
# BFS algorithm
def bfs(graph, start):
    visited, queue = set(), collections.deque([start])
    visited.add(start)
    print(f"Starting BFS from node {start}")

    while queue:
        vertex = queue.popleft()
        print(f"Visiting node {vertex}")

        for next_node in graph[vertex] - visited:
            visited.add(next_node)
            queue.append(next_node)
```

```python
# Draw the adjacency matrix
draw_adjacency_matrix(graph)
```

```python
# Draw the graph
draw_graph(graph)
```

```python
# Start BFS from node '0'
print("\nStarting BFS from node '0':")
bfs(graph, '0')
```
